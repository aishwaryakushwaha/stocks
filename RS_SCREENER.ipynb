```python
import yfinance as yf
import pandas as pd
import pandas_ta as pta


from datetime import datetime, timedelta,date
start=date.today()- timedelta(days=100)
end=date.today()

import requests
import pandas as pd
import csv
import numpy as np

def get_csv_from_url(url):
    with requests.Session() as s:
        download = s.get(url)
    
        decoded_content = download.content.decode('utf-8')
    
        cr = csv.reader(decoded_content.splitlines(), delimiter=',')
        my_list = list(cr)
        #print("Done")
    return pd.DataFrame(my_list[1:], columns = my_list[0])
```

```python
nifty_500 = get_csv_from_url("https://archives.nseindia.com/content/indices/ind_nifty500list.csv")
nifty_microcap_250= get_csv_from_url("https://archives.nseindia.com/content/indices/ind_niftymicrocap250_list.csv")

nifty_50=get_csv_from_url("https://archives.nseindia.com/content/indices/ind_nifty50list.csv")
nifty_next_50=get_csv_from_url("https://archives.nseindia.com/content/indices/ind_niftynext50list.csv")
nifty_midcap_150=get_csv_from_url("https://archives.nseindia.com/content/indices/ind_niftymidcap150list.csv")
nifty_smallcap_250=get_csv_from_url("https://archives.nseindia.com/content/indices/ind_niftysmallcap250list.csv")

nifty_50['MCap_category']="A_ NIFTY 50"
nifty_next_50['MCap_category']="B_ Nifty Next 50"
nifty_midcap_150['MCap_category']="C_ MIDCAP_150"
nifty_smallcap_250['MCap_category']="D_SmallCAP_250"
nifty_microcap_250['MCap_category']="E_Microcap_250"

nifty_750_universe=pd.concat([nifty_50,nifty_next_50,nifty_midcap_150,nifty_smallcap_250,nifty_microcap_250])
nifty_750_universe=nifty_750_universe.drop(['Series','ISIN Code'],axis=1)

nifty_750_universe['Symbol']=nifty_750_universe['Symbol']+".NS"
tickerStrings=list(nifty_750_universe['Symbol'])

nifty_750_universe.head()
```

```python
%%capture

nifty = yf.download("^NSEI",start=start,end=end,interval="1d")
nifty['rsi']=pta.rsi(nifty['Close'], timeperiod=14)


nifty500 = yf.download("^CRSLDX",start=start,end=end,interval="1d")
nifty500['rsi']=pta.rsi(nifty500['Close'], timeperiod=14)


nifty_rsi=list(nifty['rsi'])[-1]
nifty_500_rsi=list(nifty500['rsi'])[-1]


#print(nifty_rsi,nifty_500_rsi)

df_list = list()
for ticker in tickerStrings:
    data = yf.download(ticker,start=start,end=end,interval="1d")
    data['ticker'] = ticker
    data['rsi']=pta.rsi(data['Close'], timeperiod=14)
    data['ma_3mths'] = pta.sma(data["Close"], length=66)
    data['Prev_Close']=data['Close'].shift()
    data['prev_week']=data['Close'].shift(periods=5)
    data['prev_month']=data['Close'].shift(periods=22)
    data['prev_three_month']=data['Close'].shift(periods=65)
    
    
    latest=data[data.index==data.index.max()]
    df_list.append(latest)

# combine all dataframes into a single dataframe
latest_day = pd.concat(df_list)
latest_day['rs_nifty']=latest_day['rsi']/nifty_rsi
latest_day['rs_nifty_500']=latest_day['rsi']/nifty_500_rsi
latest_day=latest_day.fillna(0)
latest_day['Symbol'] = latest_day['ticker'].str.rstrip('.NS')
latest_day['1D']=(latest_day['Close']-latest_day['Prev_Close'])*100/latest_day['Prev_Close']
latest_day['1W']=(latest_day['Close']-latest_day['prev_week'])*100/latest_day['prev_week']
latest_day['1M']=(latest_day['Close']-latest_day['prev_month'])*100/latest_day['prev_month']
latest_day['3M']=(latest_day['Close']-latest_day['prev_three_month'])*100/latest_day['prev_three_month']

strong_stocks=latest_day[(latest_day['rs_nifty']>1) & (latest_day['rs_nifty_500']>1)
& (latest_day['rsi']>55) &(latest_day['Close']>latest_day['Open'])
&(((latest_day['High']-latest_day['Close'])/(latest_day['High']-latest_day['Low']))<=0.25)
&(latest_day['Close']>latest_day['Prev_Close']) & (latest_day['Close']>latest_day['ma_3mths'])]

strong_stocks=strong_stocks.reset_index()
nifty_750_universe['ticker']=nifty_750_universe["Symbol"]
strong_stocks=pd.merge(strong_stocks,nifty_750_universe[['ticker','Industry','MCap_category']], how="left",on="ticker")

out=strong_stocks[['MCap_category','Industry','Symbol','1D','1W','1M','3M']].sort_values(by=['MCap_category'],ascending=True)
out = out.round(1)
for i in ['1D','1W','1M','3M']:
   out[i]=out[i].astype(str)+"%" 

styled_html_table = out.style.set_table_styles([
    {'selector': 'caption', 'props': [('color', 'blue'), ('font-size', '16px')]},
    {'selector': 'th', 'props': [('background-color', 'lightgrey'), ('color', 'black')]},
    {'selector': 'td', 'props': [('font-size', '14px')]}
]).render()


stocks_list=out['Symbol']
stri=""
for i in stocks_list:
    stri=stri+i+","
print(stri)

from email.mime.text import MIMEText
from email.mime.application import MIMEApplication
from email.mime.multipart import MIMEMultipart
from smtplib import SMTP
import smtplib

msg = MIMEMultipart()


html =  """\
<html>
  <head></head>
  <body>
<p>
Stocks for Today <br>
""" + stri + """
<br>
    </p>
  </body>
</html>
"""

part1 = MIMEText(html, 'html')
msg.attach(part1)

html =  """\
<html>
  <head></head>
  <body>
<p>
Stocks for Today <br>
""" + styled_html_table + """
<br>
    </p>
  </body>
</html>
"""

part1 = MIMEText(html, 'html')
msg.attach(part1)

server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
server.login("majeti.koushik100@gmail.com","ezofvvrbgdgjtnah")
server.sendmail("Python_rs_list@gmail.com", "majeti.koushik100@gmail.com" , msg.as_string())
```
